### The below code is intended to introduce an example of a magic command within Jupiternotebooks. A magic command is a Python function built in the Jupiternotebook to assist users in solving frequent problems. Those can be recalled by inputting the percentage sign (%).

### The below magic command is to use matplotlib inline, which allows Jupiternotebooks to plot charts within our notebook. The code below the maplotlib inline magic command is a sniplet to generate a random scatter plot

```python
%matplotlib inline
```

```python
"""
A sample demo of a scatter plot.
"""

import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radiuses

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
